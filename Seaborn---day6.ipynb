```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
pokemon=pd.read_csv('Documents/dataset/pokemon.csv')
```

```python
pokemon
```

```python
sns.distplot(pokemon['Attack'])

#It is called univariate analysis because we are (kde-->kernel distribytion graph) considering only one column for analysis
#This plot also includes kde plot which is kernel distribution function which shows the probability of distribution 
```

```python
hist_kde=sns.distplot(pokemon['Attack'],kde=False)
hist_kde.set_title('Attack Capability')
hist_kde.set_xlabel('Attack Point')
hist_kde.set_ylabel('Frequency')
```

```python
hist_kde=sns.distplot(pokemon['Attack'],hist=False)
```

```python
sns.rugplot(pokemon['Attack'],height=0.5,axis='X')
```

```python
hist_kde=sns.distplot(pokemon['Attack'],rug=True)
hist_kde.set_title('Attack Capability')
hist_kde.set_xlabel('Attack Point')
hist_kde.set_ylabel('Frequency')
```

```python
hist_kde=sns.distplot(pokemon['Attack'],kde=False,bins=100)
```

```python
hist_kde=sns.distplot(pokemon['Attack'],kde=False,bins=100,vertical=True)
```

```python
sns.jointplot(x='Total',y='Attack',data=pokemon,xlim={300,600},ylim={50,125})
```

```python
sns.jointplot(x='Total',y='Attack', data=pokemon,kind='kde',color='r')
```

```python
sns.jointplot(x='Total',y='Attack', data=pokemon,kind='hex',color='r')
```

```python
corr_map=pokemon.corr()
corr_map
```

```python
sns.heatmap(corr_map)
```

```python

sns.heatmap(corr_map,vmax=0.7,annot=True,square=True,fmt='.2f',cmap='Reds')
```
