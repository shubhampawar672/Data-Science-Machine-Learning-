```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Apple_yield=[0.895,0.91,0.919,0.926,0.929,0.931]
```

```python
plt.plot(Apple_yield);#with semicolon we do not see unintended output
```

```python
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':15}
years=[2010,2011,2012,2013,2014,2015]
plt.plot(years,Apple_yield);
plt.title('Apple Production',fontdict=font1)
plt.xlabel('Year',fontdict=font2)
plt.ylabel('Yield (tons per hector)',fontdict=font2)
```

```python
years=range(2000,2008)
oranges=[0.851,0.861,0.89,0.912,0.926,0.933,0.938,0.942]
Apples=[0.962,0.941,0.930,0.923,0.918,0.908,0.907,0.904]
```

```python
plt.plot(years,oranges);
plt.plot(years,Apples);
plt.title('Fruit Production',fontdict=font1)
plt.xlabel('Year',fontdict=font2)
plt.ylabel('Yield (tons per hectare)',fontdict=font2)
plt.legend(['Oranges','Apples'])
```

```python
sns.set_style("dark")#"darkgrid","whitegrid","white"
plt.figure(figsize=(6,4))
plt.plot(years,oranges,marker='o');
plt.plot(years,Apples,marker='.');
plt.title('Fruit Production',fontdict=font1)
plt.xlabel('Year',fontdict=font2)
plt.ylabel('Yield (tons per hectare)',fontdict=font2)
plt.legend(['Oranges','Apples'])
```

```python
flowers_df=sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
x=flowers_df['sepal_length']
y=flowers_df['sepal_width']
plt.plot(x,y);
plt.xlabel('Sepal Length',fontdict=font2)
plt.ylabel('Sepal Width',fontdict=font2)
plt.title('Sepal Width Vs Length',fontdict=font1)
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x,y);
```

```python
plt.figure(figsize=(10,8))
plt.title('Sepal Dimensions',c='r',fontdict={'fontsize':20})
sns.scatterplot(x,y,hue=flowers_df.species,s=80);
```
